```python
import numpy as np
import pandas as pd
```

```python
df_arima_train = pd.read_csv('y_hat_df_arima_ts.csv')
df_theta_train = pd.read_csv('y_hat_df_theta_ts.csv')
```

```python
df_xgb_train = pd.read_csv('y_hat_df_xgb_ts.csv')
```

```python
df_arima_train[df_arima_train['unique_id'] == 'H1']
```

```python
df_xgb_train[df_xgb_train['unique_id'] == 'H1']
```

```python
unique_ids = df_xgb_train['unique_id'].unique()
```

```python
df_arima_train_new = df_xgb_train[['unique_id', 'ds']]
df_arima_train_new[df_arima_train_new['unique_id'] == 'H1']
```

```python
y_hat = pd.DataFrame()
for unique_id in unique_ids:

    # df_arima_train_new[df_arima_train_new['unique_id'] == unique_id]['y_hat']
    y_hat = pd.concat([y_hat, df_arima_train[df_arima_train['unique_id'] == unique_id]['y_hat']], axis=0)

y_hat.reset_index(inplace=True, drop=True)
df_arima_train_new['y_hat'] = y_hat
```
